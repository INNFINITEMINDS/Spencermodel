```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
v = np.linspace(-100, 50, 300)
```

```python
# m' = (minf-m)/mtau. Activating rate const from Spencer et al., 2012.
#'s' in 'minfs' etc. refers to *S*pencer.
def minfs(v) :
    return 1/(1-(1.11*(v+58)/(v+49))*(np.exp(-(v+49)/3)-1)/(np.exp((v+58)/20)-1)) 
def mtaus(v) :
    return 1/(0.4*((v+58)/(np.exp((v+58)/20)-1)) - 0.36*((v+49)/(np.exp(-(v+49)/3)-1)))
```

```python
# h' = (hinf-h)/htau. Inactivating rate const from Spencer et al., 2012
def hlp(v) :
    return 2.4/(1+np.exp((v+68)/3))  + 0.8/(1+np.exp(v+61.3))
def hinfs(v) :
    return hlp(v)/(hlp(v) + 3.6/(1+np.exp(-(v+21)/10)))
def htaus(v) :
    return 1/(hlp(v) + 3.6/(1+np.exp(-(v+21)/10)))
```

```python
# Activating rate const used in Rothman & Manis 2003. 
# 'r' states for *R*othman

def minfr(v) :
    return  1 / (1+np.exp(-(v + 38) / 7))    
def mtaur(v) :
    return (10 / (5*np.exp((v+60) / 18) + 36*np.exp(-(v+60) / 25))) + 0.04

# Inactivating rate const

def hinfr(v) :
    return  1 / (1+np.exp((v + 65) / 6))
def htaur(v) :
    return (100 / (7*np.exp((v+60) / 11) + 10*np.exp(-(v+60) / 25))) + 0.6
```

```python
# Dashed - Rothman's, line - Spencer's

plt.plot(v, minfs(v), 'g-', v, minfr(v), 'g--', v, hinfs(v), 'b-', v, hinfr(v), 'b--')
plt.legend(('Spencer12, m$_\infty$', 'RM03, m$_\infty$', 'Spencer12, h$_\infty$','RM03 h$_\infty$',), loc='best')
plt.show()
```

```python
# Dashed - Rothman's, line - Spencer's

plt.plot(v, mtaus(v), 'g-', v, mtaur(v), 'g--', v, htaus(v), 'b-', v, htaur(v), 'b--')
plt.show()
```

```python
# Training training
fig = plt.figure(figsize=(10,5))
fig.suptitle('Stady state activating & inactivating rate constants', fontweight = 'bold', fontsize=14)

ax = fig.add_subplot(111)
fig.subplots_adjust(top = 0.85)
ax.set_title('as functions of V')

ax.set_xlabel('voltage, mV')
ax.set_ylabel('rate constant')

ax.axis([-100, 50, 0, 1])

ax.plot(v, minfr(v), 'g-', v, minfs(v), 'g--', label='Spencer')
ax.legend()

plt.savefig('test.svg')
```
